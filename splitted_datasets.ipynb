# Import Libs and Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_first = pd.read_csv('dataset/firsthalf_v2.csv')
df_first.info()
df_first.head()
```

```python
df_second = pd.read_csv('dataset/secondhalf_v2.csv')
df_second.info()
df_second.head()
```

# One Hot Encode

Because there are still categorical features, it is necessary to treat those columns. This time, I use One Hot Encoding because there are not too many unique values, so it is still possible to be treated with OHE.

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# Categorical Features
ohe_column = ['is_holiday', 'weather_type', 'weather_description']
```

```python
encoder_first = OneHotEncoder(sparse=False)
encoder_second = OneHotEncoder(sparse=False)
```

```python
encoded_data_first = encoder_first.fit_transform(df_first[ohe_column])
encoded_data_second = encoder_second.fit_transform(df_second[ohe_column])
```

```python
encoded_data_first
```

```python
encoded_data_second
```

```python
df_encoded_first = pd.DataFrame(encoded_data_first, columns=encoder_first.get_feature_names_out())
df_encoded_second = pd.DataFrame(encoded_data_second, columns=encoder_second.get_feature_names_out())
```

```python
df_encoded_first.info()
df_encoded_second.info()
```

Simpan encoder agar dapat digunakan kembali pada pengolahan data lainnya.

```python
import joblib

# Save the encoder to a file so can be used in other files
joblib.dump(encoder_first, 'outputs/OHE_encoder_first_v2.joblib')
joblib.dump(encoder_second, 'outputs/OHE_encoder_second_v2.joblib')
```

# Concat Encoded Cols to Main df

```python
df_concated_first = pd.concat([df_first, df_encoded_first], axis=1)
df_concated_second = pd.concat([df_second, df_encoded_second], axis=1)
```

```python
df_concated_first.drop(columns=['is_holiday', 'weather_type', 'weather_description'], inplace=True)
df_concated_second.drop(columns=['is_holiday', 'weather_type', 'weather_description'], inplace=True)
```

```python
df_concated_first.info()
df_concated_first.head()
```

```python
df_concated_second.info()
df_concated_second.head()
```

# Feature Engineering

```python
df_concated_first['date_time'] = pd.to_datetime(df_concated_first['date_time'])
# Extracting date, month, and time
df_concated_first['day'] = df_concated_first['date_time'].dt.day
df_concated_first['month'] = df_concated_first['date_time'].dt.month
df_concated_first['hour'] = df_concated_first['date_time'].dt.hour

df_concated_second['date_time'] = pd.to_datetime(df_concated_second['date_time'])
# Extracting date, month, and time
df_concated_second['day'] = df_concated_second['date_time'].dt.day
df_concated_second['month'] = df_concated_second['date_time'].dt.month
df_concated_second['hour'] = df_concated_second['date_time'].dt.hour
```

```python
# Put Target column to the last column
df_target_first = df_concated_first['traffic_volume']
df_concated_first.drop(columns=['traffic_volume'], inplace=True)
df_concated_first['traffic_volume'] = df_target_first

# Put Target column to the last column
df_target_second = df_concated_second['traffic_volume']
df_concated_second.drop(columns=['traffic_volume'], inplace=True)
df_concated_second['traffic_volume'] = df_target_second
```

```python
# Display the updated DataFrame
df_concated_first.info()
df_concated_first.head()
```

```python
# Display the updated DataFrame
df_concated_second.info()
df_concated_second.head()
```

```python
df_concated_first.to_csv('dataset/df_concated_first')
df_concated_second.to_csv('dataset/df_concated_second')
```

```python
# Set the 'datetime_col' as the index of the DataFrame
df_concated_first = df_concated_first.set_index('date_time')
df_concated_second = df_concated_second.set_index('date_time')
```

```python
df_concated_first.info()
df_concated_first.head()
```

```python
df_concated_second.info()
df_concated_second.head()
```

# MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler_first = MinMaxScaler()
scaler_first = scaler_first.fit(df_concated_first)
scaled_data_first = scaler_first.transform(df_concated_first)

scaler_second = MinMaxScaler()
scaler_second = scaler_second.fit(df_concated_second)
scaled_data_second = scaler_second.transform(df_concated_second)
```

```python
import joblib

# Save the scaler to a file so can be used in other files
joblib.dump(scaler_first, 'outputs/MM_scaler_first_v2.joblib')
joblib.dump(scaler_second, 'outputs/MM_scaler_second_v2.joblib')
```

```python
df_scaled_first = pd.DataFrame(scaled_data_first, columns=scaler_first.get_feature_names_out())
df_scaled_second = pd.DataFrame(scaled_data_second, columns=scaler_second.get_feature_names_out())
```

```python
df_scaled_first.index = df_concated_first.index
df_scaled_second.index = df_concated_second.index
```

```python
df_scaled_first.info()
df_scaled_first.head()
```

```python
df_scaled_second.info()
df_scaled_second.head()
```

# Reshape to fit RNN needs

```python
trainX_first = []
trainY_first = []

trainX_second = []
trainY_second = []
```

```python
n_future = 1   # Number of data we want to look into the future based on the past data.
n_past = 24  # Number of past data we want to use to predict the future.
```

```python
for i in range(n_past, len(scaled_data_first) - n_future +1):
    trainX_first.append(scaled_data_first[i - n_past:i, 0:df_scaled_first.shape[1]])
    trainY_first.append(scaled_data_first[i + n_future - 1:i + n_future, -1])

trainX_first, trainY_first = np.array(trainX_first), np.array(trainY_first)

for i in range(n_past, len(scaled_data_second) - n_future +1):
    trainX_second.append(scaled_data_second[i - n_past:i, 0:df_scaled_second.shape[1]])
    trainY_second.append(scaled_data_second[i + n_future - 1:i + n_future, -1])

trainX_second, trainY_second = np.array(trainX_second), np.array(trainY_second)
```

```python
print('trainX_first shape == {}.'.format(trainX_first.shape))
print('trainY_first shape == {}.'.format(trainY_first.shape))

print('trainX_second shape == {}.'.format(trainX_second.shape))
print('trainY_second shape == {}.'.format(trainY_second.shape))
```

```python
np.save('outputs/train_x_first_v2.npy', trainX_first)
np.save('outputs/train_y_first_v2.npy', trainY_first)

np.save('outputs/train_x_second_v2.npy', trainX_second)
np.save('outputs/train_y_second_v2.npy', trainY_second)
```

# Baseline Model

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense, Dropout
```

```python
total_samples_first = trainX_first.shape[0]
split_index_first = int(total_samples_first * 0.9)

X_train_first = trainX_first[:split_index_first]
y_train_first = trainY_first[:split_index_first]
X_test_first = trainX_first[split_index_first:]
y_test_first = trainY_first[split_index_first:]
```

```python
total_samples_second = trainX_second.shape[0]
split_index_second = int(total_samples_second * 0.9)

X_train_second = trainX_second[:split_index_second]
y_train_second = trainY_second[:split_index_second]
X_test_second = trainX_second[split_index_second:]
y_test_second = trainY_second[split_index_second:]
```

```python
print(f'X_train_first length == {len(X_train_first)}')
print(f'X_test_first length == {len(X_test_first)}')

print(f'X_train_second length == {len(X_train_second)}')
print(f'X_test_second length == {len(X_test_second)}')
```

## LSTM

```python
model_LSTM_first = Sequential ([
    LSTM(64, activation='relu', input_shape=(trainX_first.shape[1], trainX_first.shape[2]), return_sequences=True),
    LSTM(32, activation='relu', return_sequences=False),
    Dropout(0.2),
    Dense(trainY_first.shape[1])
])

model_LSTM_first.compile(optimizer='adam', loss='mae')
model_LSTM_first.summary()
```

```python
model_LSTM_second = Sequential ([
    LSTM(64, activation='relu', input_shape=(trainX_second.shape[1], trainX_second.shape[2]), return_sequences=True),
    LSTM(32, activation='relu', return_sequences=False),
    Dropout(0.2),
    Dense(trainY_second.shape[1])
])

model_LSTM_second.compile(optimizer='adam', loss='mae')
model_LSTM_second.summary()
```

```python
# fit the model
history_LSTM_first = model_LSTM_first.fit(X_train_first, y_train_first, epochs=50, batch_size=16, validation_split=0.1, verbose=1)

plt.plot(history_LSTM_first.history['loss'], label='Training loss')
plt.plot(history_LSTM_first.history['val_loss'], label='Validation loss')
plt.legend()
```

```python
# fit the model
history_LSTM_second = model_LSTM_second.fit(X_train_second, y_train_second, epochs=50, batch_size=16, validation_split=0.1, verbose=1)

plt.plot(history_LSTM_second.history['loss'], label='Training loss')
plt.plot(history_LSTM_second.history['val_loss'], label='Validation loss')
plt.legend()
```

```python
true_value_first = df_first['traffic_volume'][split_index_first+n_past:].values
true_value_second = df_second['traffic_volume'][split_index_second+n_past:].values
```

```python
prediction_LSTM_first = model_LSTM_first.predict(X_test_first)
prediction_LSTM_second = model_LSTM_second.predict(X_test_second)
```

```python
prediction_copies_first = np.repeat(prediction_LSTM_first, df_concated_first.shape[1], axis=-1)
pred_LSTM_unscaled_first = scaler_first.inverse_transform(prediction_copies_first)[:,-1]

prediction_copies_second = np.repeat(prediction_LSTM_second, df_concated_second.shape[1], axis=-1)
pred_LSTM_unscaled_second = scaler_second.inverse_transform(prediction_copies_second)[:,-1]
```

```python
# Plot Prediction vs Real Data
plt.figure(figsize=(15,5))
plt.plot(pred_LSTM_unscaled_first[:300])
plt.plot(true_value_first[:300])
plt.title('Prediction vs Real (Unscaled)')
plt.xlabel('Datetime (harusnya)')
plt.ylabel('Traffic Volume')
plt.legend(['Prediction','True'], loc='upper left')
plt.show()
```

```python
# Plot Prediction vs Real Data
plt.figure(figsize=(15,5))
plt.plot(pred_LSTM_unscaled_second[:300])
plt.plot(true_value_second[:300])
plt.title('Prediction vs Real (Unscaled)')
plt.xlabel('Datetime (harusnya)')
plt.ylabel('Traffic Volume')
plt.legend(['Prediction','True'], loc='upper left')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error

mae_LSTM_first = mean_absolute_error(pred_LSTM_unscaled_first, true_value_first)
print(f'firsthalf: {mae_LSTM_first}')

mae_LSTM_second = mean_absolute_error(pred_LSTM_unscaled_second, true_value_second)
print(f'secondhalf: {mae_LSTM_second}')
```

## BiLSTM

```python
prediction_BiLSTM_first = model_BiLSTM_first.predict(X_test_first)
prediction_BiLSTM_second = model_BiLSTM_second.predict(X_test_second)
```
